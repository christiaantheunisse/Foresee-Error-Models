# Horizon lengths experiments

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

### Empty scenario: *traversal time*

Data for horizon lengths of: 20, 22, 24, 26 and 28 steps and a planning rate of 3 Hz.

Lengths 18 and 30 are not possible. 18 too short to see through intersection and 30 does collide with potential occluded obstacles outside the maximum Lidar range.

```python
df = pd.read_csv("horizons_vs_traversal_times.csv")
```

```python
fig, ax = plt.subplots(figsize=(5,2))

labels = [" ".join(l.split("_")[0::2]) for l in df.columns]  # ["horizon 20", ...]
labels = [l.split("_")[2] for l in df.columns]  # ["20", ...]

ax.boxplot(df.to_numpy(), labels=labels)
ax.set_xlabel("horizon length [-]")
ax.set_ylabel("traversal time[s]")
ax.set_title("Traversal times for different horizon lengths")

fig.savefig("horizons_vs_traversal_times.pdf", bbox_inches="tight")
fig.savefig("horizons_vs_traversal_times.png", bbox_inches="tight")
fig.tight_layout()
plt.show()
```
